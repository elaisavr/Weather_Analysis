# II) RAW DATAFRAME - DATA CLEANING

This notebook shows all steps done to clean first raw dataframe coming from the first part of the project - Script .py for getting the data

```python
# 1) import libraries
import pandas as pd
import numpy as np
import requests
import datetime
import sys
from datetime import date
import re
```

```python
# 2) reading raw Data Frame for cleaning 
raw_df = pd.read_csv(r'Weather_raw df.csv')
raw_df.head(10)
```

```python
# 3) delete column 'Unnamed' not useful
raw_df = raw_df.drop(raw_df.columns[0], axis=1)
```

```python
raw_df.head()
```

```python
# 4) Drop exceeding columns based on column index
raw_df = raw_df.drop(raw_df.columns[15:],axis = 1)
```

```python
raw_df.columns
```

```python
# 5) columns indexes translation from ITA to ENG
raw_df.rename(columns={'LOCALITA': 'City', 'DATA': 'Date', 'TMEDIA Â°C': 'AVG Temperature °C',\
    'TMIN Â°C':'MIN Temp °C','TMAX Â°C':'MAX Temp °C', 'PUNTORUGIADA Â°C' : 'Dew point' ,\
    'UMIDITA %':'Humidity', 'VISIBILITA km':'Visibility km', 'VENTOMEDIA km/h':'AVG Wind km/h' ,\
    'VENTOMAX km/h':'MAX Wind km/h' , 'RAFFICA km/h':'Wind gust km/h' , \
    'PRESSIONESLM mb': 'Pressure mb' , 'PRESSIONEMEDIA mb':'AVG Pressure mb' , 'PIOGGIA mm':'Rain mm' , 'FENOMENI':'Phenomenas'}, inplace=True)
```

```python
# print(raw_df.columns)
```

```python
# 6) drop not relevant rows
raw_df= raw_df.drop((raw_df[raw_df['City']!='Verona']).index)
```

```python
# raw_df.shape
```

```python
# 7) check the datas of last rows 
raw_df.tail()
```

```python
# 8) create a function to replace character ( " )
def clean_func(df):
    return df.str.replace('\"', '')

raw_df= raw_df.apply(clean_func)
raw_df.head()
```

```python
# 9) check columns data format
raw_df.info()
```

```python
# 10) change Date column format from object to date
raw_df['Date'] = pd.to_datetime(raw_df['Date'], format='%d/%m/%Y')
```

```python
# 11) check all other columns which report only numeric datas 
print(raw_df.iloc[:,2:-1].head(10))
```

```python
# 12) look for empty rows - no datas are recorded in the Weather Archive
raw_df_null = raw_df[(raw_df['AVG Temperature °C'].isnull()) & (raw_df['MIN Temp °C'].isnull()) &\
                        (raw_df['MAX Temp °C'].isnull()) & (raw_df['Dew point'].isnull()) &\
                            (raw_df['Humidity'].isnull()) & (raw_df['Visibility km'].isnull()) &\
                                (raw_df['AVG Wind km/h'].isnull()) & (raw_df['MAX Wind km/h'].isnull()) &\
                                    (raw_df['Wind gust km/h'].isnull()) & (raw_df['Pressure mb'].isnull()) &\
                                        (raw_df['AVG Pressure mb'].isnull()) & (raw_df['Rain mm'].isnull())]
```

```python
# 13) check all the indexes of the empty rows - decide not to drop the records because there won't be a right .fillna() method
raw_df_null.index
```

```python
# 14) create a function to cast all the columns with the correct dtype - parameter " errors='ignore' " is added because of the NaNs value in the DF
def clean_col(col, dtype):
    return col.str.replace(',','.', regex=False).astype(dtype , errors='ignore')
```

```python
# 15) executing columns cast - dtype 'pd.Int64Dtype' is used because Python struggles to cast into 'int64' when there are NaNs value in columns 
raw_df['AVG Temperature °C'] = clean_col(raw_df['AVG Temperature °C'], dtype='float64')
raw_df['MIN Temp °C'] = clean_col(raw_df['MIN Temp °C'], dtype='float64')
raw_df['MAX Temp °C'] = clean_col(raw_df['MAX Temp °C'], dtype='float64')
raw_df['Dew point'] = clean_col(raw_df['Dew point'], dtype='float64')
raw_df['Humidity'] = clean_col(raw_df['Humidity'], dtype='float64')
raw_df['AVG Wind km/h'] = clean_col(raw_df['AVG Wind km/h'], dtype=pd.Int64Dtype())
raw_df['MAX Wind km/h'] = clean_col(raw_df['MAX Wind km/h'],  dtype=pd.Int64Dtype())
raw_df['Wind gust km/h'] = clean_col(raw_df['Wind gust km/h'], dtype=pd.Int64Dtype())
raw_df['Pressure mb'] = clean_col(raw_df['Pressure mb'], dtype='float64')
raw_df['AVG Pressure mb'] = clean_col(raw_df['AVG Pressure mb'], dtype='float64')
raw_df['Rain mm'] = clean_col(raw_df['Rain mm'], dtype='float64')
raw_df['Visibility km'] = clean_col(raw_df['Visibility km'], dtype=pd.Int64Dtype())
```

```python
raw_df.head()
```

```python
# 16) check string values to translate
raw_df['Phenomenas'].unique()
```

```python
# 17) remove spaces at the start and at the end of all the words
raw_df['Phenomenas'] = raw_df['Phenomenas'].str.strip()
```

```python
# 18) replace with the correct translation raw_df['Phenomenas] using a dictionary and inplace
raw_df['Phenomenas'].replace({'nebbia': 'fog', 'pioggia nebbia': 'rain and fog', 'pioggia':'rain',\
     'pioggia neve nebbia':'rain snow fog', 'pioggia neve':'rain snow', 'neve nebbia':'snow fog', \
        'neve':'snow', 'pioggia temporale':'rain thunderstorm','temporale':'thunderstorm',\
            'pioggia temporale nebbia':'rain thunderstorm fog', 'temporale nebbia':'thunderstorm fog',\
                'pioggia grandine':'rain hail'}, inplace=True )
```
