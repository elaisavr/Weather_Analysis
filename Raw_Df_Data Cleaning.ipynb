```python
def clean_col(col, dtype):
    return col.str.replace(',','.', regex=False).astype(dtype , errors='ignore')
```

```python
# .astype(dtype , errors='ignore')
```

```python
raw_df['AVG Temperature °C'] = clean_col(raw_df['AVG Temperature °C'], dtype='float64')
raw_df['MIN Temp °C'] = clean_col(raw_df['MIN Temp °C'], dtype='float64')
raw_df['MAX Temp °C'] = clean_col(raw_df['MAX Temp °C'], dtype='float64')
raw_df['Dew point'] = clean_col(raw_df['Dew point'], dtype='float64')
raw_df['Humidity'] = clean_col(raw_df['Humidity'], dtype='float64')
raw_df['AVG Wind km/h'] = clean_col(raw_df['AVG Wind km/h'], dtype=pd.Int64Dtype())
raw_df['MAX Wind km/h'] = clean_col(raw_df['MAX Wind km/h'],  dtype=pd.Int64Dtype())
raw_df['Wind gust km/h'] = clean_col(raw_df['Wind gust km/h'], dtype=pd.Int64Dtype())
raw_df['Pressure mb'] = clean_col(raw_df['Pressure mb'], dtype='float64')
raw_df['AVG Pressure mb'] = clean_col(raw_df['AVG Pressure mb'], dtype='float64')
raw_df['Rain mm'] = clean_col(raw_df['Rain mm'], dtype='float64')
raw_df['Visibility km'] = clean_col(raw_df['Visibility km'], dtype=pd.Int64Dtype())
```
